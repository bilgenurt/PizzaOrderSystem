```python

class Pizza:
    def __init__(self, description,prices):
        self.description = description
        self.prices = prices
    
    def get_description(self):
        return self.description

    def get_cost(self):
        return self.prices
```

```python

class KlasikPizza(Pizza):
    def __init__(self):
        super().__init__("Klasik pizza",50)


class MargheritaPizza(Pizza):
    def __init__(self):
        super().__init__("Margherita pizza",30)


class TurkPizza(Pizza):
    def __init__(self):
        super().__init__("Turk pizza",40)


class DominosPizza(Pizza):
    def __init__(self):
        super().__init__("Dominos pizza",45)
```

```python

class Decorator(Pizza):
    def __init__(self, pizza):
        self.pizza = pizza

    def get_description(self):
        return self.pizza.get_description()

    def get_cost(self):
        return self.pizza.get_cost()
```

```python

class ZeytinSosu(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 3
        self.description = " Zeytin Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class MantarSosu(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 8
        self.description = " Mantar Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class KeciPeyniriSosu(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 7
        self.description = " Keçi Peyniri Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class EtSosu(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 10
        self.description = " Et Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class SoganSosu(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 4
        self.description = " Soğan Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class MisirSosu(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 5
        self.description = " Mısır Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price
```

```python

import csv
import os
from datetime import datetime

def data(name, id_no, card_no, description, card_password, total_cost):
    
    time = datetime.now()
    order_time = time.strftime("%Y-%m-%d %H:%M:%S")

    
    orders = [name, id_no, card_no, description, order_time, card_password, total_cost]

    files = os.path.isfile("Orders_Database.csv")

    
    with open("Orders_Database.csv", mode="a", newline="") as file:
        writer = csv.writer(file)
        if files:
            writer.writerow(orders)
        else:
            od = (["Name", "Id", "Card Number", "Order Description", "Order Time", "Card Password", "Total Price"])
            writer.writerow(od)
            writer.writerow(orders)
```

```python

def main():
  
  with open("Menu.txt", "w") as f:
    f.write("* Lütfen Bir Pizza Tabanı Seçiniz:\n1: Klasik Pizza -- İçindekiler => mantar, domates, sucuk, salam, sosis -- Fiyat => 50 TL \n2: Margarita Pizza -- İçindekiler => domates, mozarella, fesleğen -- Fiyat => 30 TL \n3: Türk Pizza -- İçindekiler => kıyma, biber, domates -- Fiyat => 40 TL\n4: Sade Pizza -- İçindekiler => peynir, biber, domates, sosis -- Fiyat => 45 TL\n")
    f.write("----------------------------------------------------------------------------------\n")
    f.write("* ve seçeceğiniz sos:\n11: Zeytin Sosu -- Fiyat => 3 TL\n12: Mantar Sosu -- Fiyat => 8 TL\n13: Keçi Peyniri Sosu -- Fiyat => 7 TL\n14: Et Sosu -- Fiyat => 10 TL\n15: Soğan Sosu -- Fiyat => 4 TL\n16: Mısır Sosu -- Fiyat => 5 TL\n")

  with open("Menu.txt", "r") as f:
    content = f.read()
    print(content)


    
    pizza_choice = input("Lütfen pizza seçiniz (1-4): ")
    while pizza_choice not in ["1", "2", "3", "4"]:
        pizza_choice = input("Lütfen geçerli bir pizza numarası seçiniz (1-4): ")

    
    if pizza_choice == "1":
        pizza = KlasikPizza()
    elif pizza_choice == "2":
        pizza = MargheritaPizza()
    elif pizza_choice =="3":
        pizza = TurkPizza()
    else:
        pizza = DominosPizza()

   

    
    sos_choice = input("Lütfen sos seçiniz (11-16): ")
    while sos_choice not in ["11", "12", "13", "14", "15", "16"]:
        sos_choice = input("Lütfen geçerli bir sos harfi seçiniz (11-16): ")
    
    if sos_choice == "11":
        sos = ZeytinSosu(pizza)
    elif sos_choice == "12":
        sos = MantarSosu(pizza)
    elif sos_choice == "13":
        sos = KeciPeyniriSosu(pizza)
    elif sos_choice == "14":
        sos = EtSosu(pizza)
    elif sos_choice == "15":
        sos = SoganSosu(pizza)
    elif sos_choice == "16":
        sos = MisirSosu(pizza)
    

  
   
    name = input("Lütfen Adınızı giriniz: ")
    
    id_no = input("Lütfen TC Kimlik Numaranızı giriniz: ")

    card_no = input("Lütfen Kredi Kartı Numaranızı giriniz: ")
      
    card_password = input("Lütfen Kredi Kartı Şifrenizi giriniz: ")



    
    order_description =  pizza.get_description() + " ile" + sos.get_description()

    
    cost = pizza.get_cost() + sos.get_cost()

    
    data(name, id_no, card_no, order_description, card_password,cost)

    
    print("\nSipariş Detayları:")
    print("------------------------------------")
    print("Ad: ", name)
    print("TC Kimlik Numarası: ", id_no)
    print("Kredi Kartı Numarası: ", card_no)
    print("Sipariş Açıklaması: ", order_description)
    print("Toplam Fiyat: ", cost , "TL")
    print("Sipariş Zamanı: ", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    print("\nSiparişiniz onaylanmıştır. Teşekkür ederiz!")
    
    


```
